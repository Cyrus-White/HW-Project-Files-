```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from sklearn.cluster import KMeans

# Load the image
image_path = '/content/' + 'Tesla.png'
image = io.imread(image_path)

# Reshape the image to a 2D array of RGB pixels
image_reshaped = image.reshape((-1, 3))

# Apply K-means clustering
#kmeans = KMeans(n_clusters=5, random_state=42)  # Change n_clusters for more or fewer regions
#kmeans.fit(image_reshaped)

# Get the segmented image by replacing each pixel with its corresponding cluster center
#segmented_image = kmeans.cluster_centers_[kmeans.labels_].reshape(image.shape)

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(image_reshaped)  # Fit the KMeans model to your data
segmented_image = kmeans.cluster_centers_[kmeans.labels_].reshape(image.shape)

# Plot the original and segmented image
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Original image
axes[0].imshow(image)
axes[0].set_title('Original Image')
axes[0].axis('off')

# Segmented image
axes[1].imshow(segmented_image.astype(int))
axes[1].set_title('Segmented Image')
axes[1].axis('off')

plt.show()


```
